```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_return(df):
    for i in range(1,10):
        return_data = df['t'+str(i)+'_ret_on_t1open'].values
        plt.figure(figsize=(15, 7))
        plt.plot(return_data)
        plt.show()

if __name__ == "__main__":
    
    stock_price = pd.read_csv("../data/panel_data/ema100.csv", index_col = 0)
    plot_return(stock_price)
```
